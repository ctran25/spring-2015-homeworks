# Data Parsing and Representation:

### Convert Textfiles to DataFrames

#### To see dataframes, un-comment last line of each of the following 3 code cells

* TextFile: allr_recipes.txt
* DataFrame: r_dataFrame

```python
recipes = open("allr_recipes.txt", 'r')

dictionary = {}  # dictionary of Countries, the ingredients, the number of times the ingredient occurs
counts ={}       # dictionary of Countries and how many recipes are included for each

for entry in recipes:
    line = entry.split()
    country = line[0]
    ingrediants = line[1:]
    if country not in dictionary:
        dictionary[country] = {}
        counts[country] = 1
    else:
        counts[country] += 1
    for food in ingrediants:
        if food in dictionary[country]:
            dictionary[country][food] += 1
        else:
            dictionary[country][food] = 1

r_dataFrame = pd.DataFrame(dictionary) # create the DataFrame
r_dataFrame = r_dataFrame.fillna(0)    # fill all NaN with 0
labels = r_dataFrame.columns           # countries

rr = r_dataFrame   # create copy of DataFrame

# normalize number entries
# divide values of each ingredient for each country by dividing the values by the number
# of recipes each country has in this dataset
sizeCol = len(r_dataFrame.columns)  
cNumber = 0
for column in range(sizeCol):
    cName = r_dataFrame.columns[cNumber]
    r_dataFrame[cName] = r_dataFrame[cName]/counts[cName]
    cNumber += 1

# Transpose the DataFrame we already have
r_dataFrame = r_dataFrame.T
# r_dataFrame
```

* TextFile: epic_recipes.txt
* DataFrame: e_dataFrame

```python
e_recipes = open("epic_recipes.txt", 'r')

e_dictionary = {}
e_counts ={}
for entry in e_recipes:
    line = entry.split()
    country = line[0]
    ingrediants = line[1:]
    if country not in e_dictionary:
        e_dictionary[country] = {}
        e_counts[country] = 1
    else:
        e_counts[country] += 1
    for food in ingrediants:
        if food in e_dictionary[country]:
            e_dictionary[country][food] += 1
        else:
            e_dictionary[country][food] = 1        
            
e_dataFrame = pd.DataFrame(e_dictionary)
e_dataFrame = e_dataFrame.fillna(0)

e_labels = e_dataFrame.columns

e_sizeCol = len(e_dataFrame.columns)
e_cNumber = 0
for column in range(e_sizeCol):
    e_cName = e_dataFrame.columns[e_cNumber]
    e_dataFrame[e_cName] = e_dataFrame[e_cName]/e_counts[e_cName]
    e_cNumber += 1

e_dataFrame = e_dataFrame.T
# e_dataFrame
```

* TextFile: menu_recipes.txt
* DataFrame: m_dataFrame

```python
m_recipes = open("menu_recipes.txt", 'r')

m_dictionary = {}
m_counts ={}
for entry in m_recipes:
    line = entry.split()
    country = line[0]
    ingrediants = line[1:]
    if country not in m_dictionary:
        m_dictionary[country] = {}
        m_counts[country] = 1
    else:
        m_counts[country] += 1
    for food in ingrediants:
        if food in m_dictionary[country]:
            m_dictionary[country][food] += 1
        else:
            m_dictionary[country][food] = 1        
            
m_dataFrame = pd.DataFrame(m_dictionary)
m_dataFrame = m_dataFrame.fillna(0)

m_labels = m_dataFrame.columns

m_sizeCol = len(m_dataFrame.columns)
m_cNumber = 0
for column in range(m_sizeCol):
    m_cName = m_dataFrame.columns[m_cNumber]
    m_dataFrame[m_cName] = m_dataFrame[m_cName]/m_counts[m_cName]
    m_cNumber += 1

m_dataFrame = m_dataFrame.T
# m_dataFrame
```
