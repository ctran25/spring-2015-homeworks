## Data Parsing and Representation:

### Convert Textfiles to DataFrames

* TextFile: allr_recipes.txt
* DataFrame: r_dataFrame

```python
recipes = open("allr_recipes.txt", 'r')

dictionary = {}
counts ={}
for entry in recipes:
    line = entry.split()
    country = line[0]
    ingrediants = line[1:]
    if country not in dictionary:
        dictionary[country] = {}
        counts[country] = 1
    else:
        counts[country] += 1
    for food in ingrediants:
        if food in dictionary[country]:
            dictionary[country][food] += 1
        else:
            dictionary[country][food] = 1  
            
r_dataFrame = pd.DataFrame(dictionary)
r_dataFrame = r_dataFrame.fillna(0)
labels = r_dataFrame.columns

rr = r_dataFrame

sizeCol = len(r_dataFrame.columns)
cNumber = 0
for column in range(sizeCol):
    cName = r_dataFrame.columns[cNumber]
    r_dataFrame[cName] = r_dataFrame[cName]/counts[cName]
    cNumber += 1

r_dataFrame = r_dataFrame.T
```
