```python
%matplotlib inline
import requests
import json
import pandas as pd
from matplotlib import pyplot as plt
```

```python
exoplanets = pd.read_csv('exoplanets.csv')
```

```python
confirmed_exoplanets = exoplanets[exoplanets.Planet_Name.notnull()]
```

```python
print  exoplanets.shape
```

```python
confirmed_exoplanets
```

```python
print confirmed_exoplanets.shape
print confirmed_exoplanets.columns
```

```python
confirmed_exoplanets = confirmed_exoplanets.drop('t0_uncer_days', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('e_Period_days', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('Depth_ppm', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('e_d/R*', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('e_r/R*', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('b', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('e_b', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('SNR', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('chi', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('log(g)_cm/s2', 1)
confirmed_exoplanets = confirmed_exoplanets.drop('f_Teff', 1)
```

```python
confirmed_exoplanets = confirmed_exoplanets.rename(columns={'Kepler ID': 'KeplerID', 't0_days': 'timeOfTransitCenter_days', 'a_AU': 'semiMajorAxis_AU', 'Teq_K': 'eq_surfaceTemp_K', 'd/R*': 'sep_starRadius', 'r/R*':'planetR_starR', 'Teff_K':'starEffTemp_K'})
confirmed_exoplanets = confirmed_exoplanets.fillna(0)
confirmed_exoplanets
```

```python
print confirmed_exoplanets.shape
print confirmed_exoplanets.columns
```

```python
print "AVERAGE: ============================================"

keplar_mag = confirmed_exoplanets.Kepmag.mean()
print "keplar magnitude                = ", keplar_mag

timeTransitCenter = confirmed_exoplanets.timeOfTransitCenter_days.mean()
print "time of transit center (days)   = ", timeTransitCenter

average_period = confirmed_exoplanets.Period_days.mean()
print "planetary period (days)         = ", average_period

planet_meanRadius = confirmed_exoplanets.p_Rad_Earth.mean()
print "planetary radius (Earth radii)  = ", planet_meanRadius

average_semiMajorAxis = confirmed_exoplanets.semiMajorAxis_AU.mean()
print "semi-major axis (AU)            = ", average_semiMajorAxis

average_eq_temp = confirmed_exoplanets.eq_surfaceTemp_K.mean()
print "equilibrium temperature (K)     = ", average_eq_temp

average_transit_time = confirmed_exoplanets.Dur_h.mean()
print "transit duration (hours)        = ", average_transit_time

average_sep_radius = confirmed_exoplanets.sep_starRadius.mean()
print "ratio: separation/star radius   = ", average_sep_radius

average_planetR_starR = confirmed_exoplanets.planetR_starR.mean()
print "ratio: planetRadius/starRadius  = ", average_planetR_starR

average_star_temp = confirmed_exoplanets.starEffTemp_K.mean()
print "effective star temp (K)         = ", average_star_temp

average_star_radius = confirmed_exoplanets.s_Rad_solar.mean()
print "ratio: separation/star radius   = ", average_star_radius
```
