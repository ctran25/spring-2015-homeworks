## Hypothesis 2: Hierarchically clustered stars will have unique numbers of planets [FALSE]

First, find the average of the 6 attributes for normalization. Do not include 0's in the average.

```python
star_cols = ['star_dist_pc', 'star_eff_temp', 'star_mass', 'star_radius', 'star_density', 'star_age']
for col in star_cols:
    count = 0
    total = 0
    for index, row in stars.iterrows():
        if row[col] != 0:
            count = count + 1
            total = total + row[col]
    average = total/count
    print 'count = ', count
    print 'total = ', total
    print 'average ', col, ' = ', average
    print 
```

```python
avg_distance = 418.856351744
avg_effTemp = 5583.43063063
avg_starMass = 1.01796407186
avg_starRad = 2.11750620347
avg_starDen = 2.30847877095
avg_starAge = 4.26997275204
```

```python
def avgDist(x):
    return x['star_dist_pc'] / avg_distance
def avgTemp(x):
    return x['star_eff_temp'] / avg_effTemp
def avgMass(x):
    return x['star_mass'] / avg_starMass
def avgRad(x):
    return x['star_radius'] / avg_starRad
def avgDen(x):
    return x['star_density'] / avg_starDen
def avgAge(x):
    return x['star_age'] / avg_starAge

starNormal = stars
starNormal['star_dist_pc'] = starNormal.apply(avgDist, axis=1)
starNormal['star_eff_temp'] = starNormal.apply(avgTemp, axis=1)
starNormal['star_mass'] = starNormal.apply(avgMass, axis=1)
starNormal['star_radius'] = starNormal.apply(avgRad, axis=1)
starNormal['star_density'] = starNormal.apply(avgDen, axis=1)
starNormal['star_age'] = starNormal.apply(avgAge, axis=1)

starNormal = starNormal.drop(['spectral_type', 'planet_name'], axis=1, inplace=False)
starNormal = starNormal[starNormal.star_dist_pc != 0]
starNormal = starNormal[starNormal.star_eff_temp != 0]
starNormal = starNormal[starNormal.star_mass != 0]
starNormal = starNormal[starNormal.star_radius != 0]
starNormal = starNormal[starNormal.star_density != 0]
starNormal = starNormal[starNormal.star_age != 0]

starNormal2 = starNormal.set_index('star_name')
starNormal2.index.name = None
```

### Hierarchically cluster with euclidean distances and a threshold of 0.8 to create 5 clusters for 1-5 possible planets as (6 and 7 planets are rare and have been filtered out due to insufficient measurements above)

```python
Z = hr.linkage(starNormal2, method='complete', metric='euclidean')
print Z.shape
labels = starNormal2.index
distanceMatrix = pdist(starNormal2,'euclidean')
```

```python
fig = plt.figure(figsize=(100,40))
T = hr.dendrogram(Z, leaf_font_size=35, color_threshold = 0.8,labels=labels)
fig.show()
```
